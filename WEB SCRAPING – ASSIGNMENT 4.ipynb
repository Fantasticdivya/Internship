```python
#Creating empty lists and then finding details of most viewed youtube vidoes
rows = driver.find_elements(By.XPATH, '/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[1]/tbody/tr')
details = []
ranks=[]
name=[]
artist=[]
views=[]
upload=[]
date=[]
rank = 1  
for row in rows:
    cells = row.find_elements(By.XPATH, 'td')
    ranks.append(rank)
    name.append(cells[0].text)
    artist.append(cells[1].text)
    views.append(cells[2].text)
    date.append(cells[3].text)
    rank += 1
```

```python
# Checking length of lists to make dataframe
print(len(ranks),len(name),len(artist),len(views))
```

```python
#creating dataframe
df=pd.DataFrame({})
df["Rank"]=ranks
df["Name"]=name
df["Artist"]=artist
df['Upload date']=date
df['Views']=views
df
```
