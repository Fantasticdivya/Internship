# Question-1 Write a python program which searches all the product under a particular product from www.amazon.in. The
product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for
guitars.

```python
#take user input and search for the desired product
def product_name():
    name=input("Enter the product name")
    return name

driver = webdriver.Chrome()
driver.get("https://www.amazon.in")

search=driver.find_element(By.XPATH, "/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input")
search.send_keys(product_name())

submit_btn=driver.find_element(By.XPATH, "/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input")
submit_btn.click()
    
time.sleep(3)
```

```python
# Scrape all product Urls

product_urls = []
start = 0
end = 3

##looping for scrapping 3 pages
for page in range(start,end): 
    url=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url:
        product_urls.append(i.get_attribute("href"))   #appending the urls
    next_btn=driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')   
    next_btn.click()
    time.sleep(3)
```

```python
len(product_urls)
```

```python
#creating empty lists and then fetching and stroing data in that list
Brand =[]
Name = []
Price =[]
Return =[]
Availability =[]
Expected_delivery =[]


for url in product_urls:
    driver.get(url)
    time.sleep(2)
    
    try:
        brand_tag=driver.find_element(By.XPATH,'//span[@class="a-size-small tabular-buybox-text-message"]/a')
        Brand.append(brand_tag.text)
    except NoSuchElementException:
        Brand.append('-')
        
    try:
        name_tag=driver.find_element(By.XPATH,'//span[@class="a-size-large product-title-word-break"]')
        Name.append(name_tag.text)
    except NoSuchElementException:
        Name.append('-')
        
    try:
        price_tag=driver.find_element(By.XPATH,'//span[@class="a-price-whole"]')
        Price.append(price_tag.text)
    except NoSuchElementException:
        Price.append('-')
        
    try:
        return_tag=driver.find_element(By.XPATH,'//div[@id="RETURNS_POLICY"]/span/div[2]/span')
        Return.append(return_tag.text)
    except NoSuchElementException:
        Return.append('-')    
        
    try:
        avail_tag=driver.find_element(By.XPATH,'//span[@class="a-size-medium a-color-success"]')
        Availability.append(avail_tag.text)
    except NoSuchElementException:
        Availability.append('-')        
        
    try:
        delivery_tag=driver.find_element(By.XPATH,'//div[@class="a-spacing-base"]/span/span')
        Expected_delivery.append(delivery_tag.text)
    except NoSuchElementException:
        Expected_delivery.append('-')     
    
```

```python
#making dataframe
df = pd.DataFrame({'Name' : Name, 'Brand' : Brand, 'Price' : Price, 'Return/Exchange' : Return, 'Availiblity' : Availability, 'Delivery' : Expected_delivery, 'URL': product_urls })
df
```

```python
#Saving the Dataframe as csv file
df.to_csv('Guitar.csv')
```

# Question-7 Write a python program to scrape the details for all billionaires from www.forbes.com. Details to be scrapped: “Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, “Industry”.

```python
options = Options()
options.add_argument("--start-maximized") 
```

```python
rank = []
name = []
net_worth = []
age = []
citizenship = []
source = []
industry = []
```

```python
def extract_data():
   
    rows = driver.find_elements(By.XPATH, '//*[@role="row"]')       #Each row data

    time.sleep(3)

    for row in rows:
        row_record = list(row.text.split('\n'))
        rank.append(row_record[0][:-1])
        name.append(row_record[1])
        net_worth.append(row_record[2][0:-2])
        age.append(row_record[3])
        citizenship.append(row_record[4])
        source.append(row_record[5])
        industry.append(row_record[6])

    try:
        next_button = driver.find_element(By.XPATH, '//*[@class="t7aCs7DQ _0BYTGi5u TNAO5JyH bMlalnxk _5nHoWK7c"][2]')
        next_button.click()

        time.sleep(3)
        extract_data()

    except:
        return
```

```python
driver = webdriver.Chrome(service=Service(), options=options) #Web driver Setup

url = "https://www.forbes.com/billionaires/" 
driver.get(url)

extract_data()

driver.quit()
```

```python
# creating dictionary
billionaires_data = {
    'Rank': rank,
    'Name': name,
    'Net Worth': net_worth,
    'Age': age,
    'Citizenship': citizenship,
    'Source': source,
    'Industry': industry
}
```

```python
billionaires_df = pd.DataFrame(billionaires_data)   #Create Dataframe
```

```python
billionaires_df = billionaires_df[billionaires_df['Rank']!='RAN']   #Filtering out extra data
```

```python
billionaires_df = billionaires_df.reset_index()
```

```python
billionaires_df = billionaires_df.drop(columns=['index'])
```
