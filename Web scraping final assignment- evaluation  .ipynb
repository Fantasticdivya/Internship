# Q-1 In this question you have to scrape data using the filters available on the webpage You have to use the location and salary filter.
You have to scrape data for “Data Scientist” designation for first 10 job results.
You have to scrape the job-title, job-location, company name, experience required.
The location filter to be used is “Delhi/NCR”. The salary filter to be used is “3-6” lakhs

```python
#importing required libraries
import selenium
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
import pandas as pd
```

```python
#opening webdriver automically
driver= webdriver.Chrome()
```

```python
#Opening website using web browser
driver.get('https://www.naukri.com/')
```

```python
#By using absolute xpath searching for data scientist jobs
designation= driver.find_element(By.CLASS_NAME,"suggestor-input ")
designation.send_keys('Data Scientist')
```

```python
# clicking search button by using click command using absolute xpath
search = driver.find_element(By.CLASS_NAME,"qsbSubmit")
search.click()
```

```python
#using filter to fetch the jobs from Delhi/NCR
location = driver.find_element(By.XPATH,"/html/body/div/div/main/div[1]/div[1]/div/div/div[2]/div[5]/div[2]/div[3]/label/p/span[1]")
location.click()
```

```python
# fetching the salary range of 3-6 lakhs using salary filter
salary_range = driver.find_element(By.XPATH,"/html/body/div/div/main/div[1]/div[1]/div/div/div[2]/div[5]/div[2]/div[2]/label/p/span[1]")
salary_range.click()
```

```python
#Creating enpty lists
job_title =[]
job_location = []
company_name = []
experience_required = []
```

```python
#Scraping all the fields for which we created empty lists above by using relative xpath and using append method to store all the data in the creatde list
title_tags = driver.find_elements(By.XPATH,'//a[@class="title "]')
for i in title_tags[0:10]:
    title = i.text
    job_title.append(title)
    
    
location_tags = driver.find_elements(By.XPATH,'//span[@class="locWdth"]')
for i in location_tags[0:10]:
    location = i.text
    job_location.append(location)
    
company_tags = driver.find_elements(By.XPATH,'//a[@class=" comp-name mw-25"]') 
for i in company_tags[0:10]:
    company = i.text
    company_name.append(company)   

    
experience_tags = driver.find_elements(By.XPATH,'//span[@class="expwdth"]') 
for i in experience_tags[0:10]:
    experience = i.text
    experience_required.append(experience)      
```

```python
len(job_title), len(job_location), len(company_name), len(experience_required)
```

```python
#creating dataframe fpr top 10 job results
df = pd.DataFrame({'Title': title_tags[0:10],'Location': job_location[0:10], 'Company name': company_name[0:10], 'Experience': experience_required[0:10]})
df
```

# Q3: Scrape 100 reviews data from flipkart.com for iphone11 phone. You have to go the link: https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART
As shown in the above page you have to scrape the tick marked attributes. These are:
1. Rating
2. Review summary
3. Full review
4. You have to scrape this data for first 100reviews

```python
#connecting to the webdriver
driver = webdriver.Chrome()
```

```python
#opening the page where i-phone reviews are shown
driver.get("https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART")
```

```python
#creating empty list for saving data
rating= []
review_summary = []
full_review = []
```

```python
#scraping top 100 rating, review summary and full reviews and then append then to list empty list created
start = 0
end = 10
for page in range(start,end):
    ratings =  driver.find_elements(By.XPATH,'//div[@class="XQDdHH Ga3i8K"]')
    for i in ratings[0:100]:
        rating.append(i.text)
        
for page in range(start,end):
    rev = driver.find_elements(By.XPATH,'//p[@class="z9E0IG"]')
    for i in rev[0:100]:
        review_summary.append(i.text)
        
for page in range(start,end):
    full_rev = driver.find_elements(By.XPATH,'//div[@class="ZmyHeo"]')
    for i in full_rev[0:100]:
        full_review.append(i.text)
        
        
       
      
next_button = driver.find_elements(By.XPATH, '/html/body/div/div/div[3]/div/div/div[2]/div[13]/div/div/nav/a[11]/span')
next_button.click()
time.sleep(5)

```

```python
#checking length of lists to make dataframe
len(rating),len(review_summary),len(full_review)
```

```python
#making dataframe
df2=pd.DataFrame({"Rating":rating[0:100],"Review Summary": review_summary[0:100],"Full Review": full_review[0:100]})
df2
```

# Q6: Write a python program to scrape data for Top 1000 Quotes of All Time.
The above task will be done in following steps:
1. First get the webpagehttps://www.azquotes.com/
2. Click on Top Quote
3. Than scrap a) Quote b) Author c) Type Of Quotes

```python
#opening webdriver
driver = webdriver.Chrome()
```

```python
#opening website of azquotes.com through driver
driver.get('https://www.azquotes.com/')
```

```python
#clicking on top quotes button through click command
top_quotes = driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')
top_quotes.click()
```

```python
#creating empty lists 
quote=[]
author=[]
type_of_quote=[]
```

```python
#fething the quotes,author and type of qutes using xpath and then appending it to the empty list and then for scraping 1000 quotes navigating through pages using next button
start=0
end=10
for page in range(start,end):
    qut= driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in qut:
        quote.append(i.text)
        
for page in range(start,end):
    auth= driver.find_elements(By.XPATH,'//div[@class="author"]')
    for d in auth:
        author.append(d.text)
        
for page in range(start,end):
    type= driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for d in type:
        type_of_quote.append(d.text)
        
        
next_button2= driver.find_element(By.XPATH,"/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[3]/li[13]/a")
next_button2.click()
time.sleep(3)
```

```python
#making dataframe
df4= pd.DataFrame({"Quote":quote[0:1000],"Author":author[0:1000],"Type Of Quote":type_of_quote[0:1000]})
df4
```

# Q8: Write a python program to display list of 50 Most expensive cars in the world (i.e. Car name and Price) from https://www.motor1.com/
This task will be done in following steps:
1. First get the webpage https://www.motor1.com/
2. Then You have to type in the search bar ’50 most expensive cars’
3. Then click on 50 most expensive cars in the world..
4. Then scrap the mentioned data and make the dataframe.

```python
#connecting to driver
driver = webdriver.Chrome()
```

```python
#opening the website of motor1.com through driver
driver.get('https://www.motor1.com/')
```

```python
#serching the cars by using send key
search = driver.find_element(By.XPATH, "/html/body/div[9]/div[6]/form/input[2]")
search.send_keys('50 most expensive cars')
```

```python
#Searching the cars using click command
cars= driver.find_element(By.XPATH,'/html/body/div[9]/div[6]/form/input[2]')
cars.click()
```

```python
# creating empty list
car_name =[]
car_price= []
```

```python
# scraping details of top 50 cars using xpath and appending it in list
carname= driver.find_elements(By.XPATH,'//h3[@class="subheader"]')
for i in carname[0:50]:
    car_name.append(i.text)
    
    
carprice = driver.find_elements(By.TAG_NAME,"strong") 
for i in carprice[0:50]:
    car_price.append(i.text)
    
```

```python
#checking length to make dataframe
len(car_name),len(car_price)
```

```python
#making dataframe
df4 = pd.DataFrame({'Car Name': car_name[:50], 'car_price': car_price[:50]})
df4
```
